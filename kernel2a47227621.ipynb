```r
x<-c(1,5,90,76,34,43,67,33,56,76)
cat(x,'\n')
cat('Mean of x',mean(x),'\n')
cat('Median of x',median(x),'\n')
cat('Variance of x',var(x),'\n')
cat('Standard Deviation of x',sd(x),'\n')
```

```r
print(x)
print(paste('Mean of x',mean(x)))
print(paste('Median of x',median(x)))
print(paste('Variance of x',var(x)))
print(paste('Standard Deviation of x',sd(x)))
```
